```python
import numpy as np
matrix=np.array([[3,4],[0,5]])

tr = matrix.T
print(tr)
transpose = np.transpose(matrix)
print(matrix)
print(transpose)
det = np.linalg.det(matrix)
print(det)
detTasnpose = np.linalg.det(transpose)
print(detTasnpose)


dotProductOfAATraspose = matrix.T@matrix
print(dotProductOfAATraspose)

eigvalues, eigvectors = np.linalg.eig(matrix)
print(eigvalues)
print(eigvectors)
```

***CLass 2: Matrix Operations***

```python
import numpy as np
A = np.array([[3,4],[0,5]])
B = np.transpose(A)
print(B)

C = A@B
print(C)

D=B@A
print(D)

eigvalues, eigectors = np.linalg.eig(C)
print(eigvalues)
print(eigectors)
```
